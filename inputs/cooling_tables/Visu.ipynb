```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# File
table = open('gnat-sternberg.cooling','r')
lines = table.readlines()

# Lists
logT = []
logL = []

for line in lines[8:]:
    
    split = line.split()
    
    logT.append(float(split[0]))
    logL.append(float(split[1]))

# Close
table.close()

# Plot
fig,ax = plt.subplots()
ax.plot(logT,logL)
ax.set_xlabel('log T')
ax.set_ylabel('log $\Lambda$')
ax.set_ylim(-25,-21)
```
